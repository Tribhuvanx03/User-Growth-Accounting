```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
wau_data_csv = pd.read_csv('/content/WAU.csv')
```

```python
wau_sets = [set(wau_data_csv[col].dropna()) for col in wau_data_csv.columns]

for i, wau_set in enumerate(wau_sets[:5]):
    print(f"Week {i+1}:", wau_set)
```

```python
len(wau_sets[2])
```

### **Begin the Data Analysis**

```python
# Lists to store results
new_users = []
retained_users = []
resurrected_users = []
churned_users = []

# All previously seen users
seen_users = set(wau_sets[0])

# Iterate through the weeks to calculate the metrics
for i in range(1, len(wau_sets)):
    current_week = wau_sets[i]
    previous_week = wau_sets[i-1]

    # Calculate new users
    new = len(current_week - seen_users)
    # Calculate retained users
    retained = len(current_week & previous_week)
    # Calculate resurrected users
    resurrected = len(current_week & seen_users - previous_week)
    # Calculate churned users
    churned = len(previous_week - current_week)

    new_users.append(new)
    retained_users.append(retained)
    resurrected_users.append(resurrected)
    churned_users.append(churned)

    # Update the seen users set
    seen_users.update(current_week)

# Storing the results
wau_new = pd.DataFrame({
    'Week': wau_data_csv.columns[1:],
    'New Users': new_users,
    'Retained Users': retained_users,
    'Resurrected Users': resurrected_users,
    'Churned Users': churned_users
})

# Quick Ratio
wau_new['Quick Ratio'] = (wau_new['New Users'] + wau_new['Resurrected Users']) / wau_new['Churned Users']

print(wau_new)
```

```python
l = []
for i in range(2, 57):
  l.append(f'w{i}')
```

###  **Graph Plotting**


```python
fig, ax1 = plt.subplots(figsize=(14, 8))

# Plot stacked bars for New, Resurrected, and Churned Users
p1 = ax1.bar(l, wau_new['New Users'], label='new', color='blue')
p2 = ax1.bar(l, wau_new['Resurrected Users'], bottom=wau_new['New Users'], label='resurrected', color='red')
p3 = ax1.bar(l, (-1)*wau_new['Churned Users'],label='churned', color='green')

# Plot the Quick Ratio
ax2 = ax1.twinx()
ax2.plot(l, wau_new['Quick Ratio'], label='(new+res.)/churned', color='purple', marker='o')

# Labels and Titles
ax1.set_xlabel('Week')
ax1.set_ylabel('Number of Users')
ax2.set_ylabel('Ratio')
plt.title('WAU Growth Accounting')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.setp(ax1.get_xticklabels(), rotation=45, ha="right")
plt.grid(True)
plt.tight_layout()

plt.show()
```
